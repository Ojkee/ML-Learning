```python
def mse(yhs, ys):
    return sum((yh - y) ** 2 for yh, y in zip(yhs, ys)) / len(ys)
```

```python
xs = [
    [2, 3, -1],
    [3, -1, -0.5],
    [0.5, 1, 1],
    [1, 1, -1],
]
ys = [1, -1, -1, 1]
```

```python
EPOCHS = 50

for i in range(EPOCHS):
    nn.zero_grad()

    preds = [nn(x)[0] for x in xs]
    loss = mse(preds, ys)
    loss.backward()
    print(f"{i}: {str(loss):>12}")
    nn.train(learning_rate=0.1)
```
